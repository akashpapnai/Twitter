```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'Data.csv')
data.head(5)
```

```python
tweet_created_at = data.tweet_created_at
tweet_created_at_df = pd.DataFrame(data=tweet_created_at.value_counts())
tweet_created_at_df.columns = ['Total']
```

```python
tweet_created_at_df
```

```python
tweet_created_on_holiday_bool = data.tweet_created_on_holiday_bool
tweet_created_on_holiday_bool_df = pd.DataFrame(tweet_created_on_holiday_bool.value_counts())
tweet_created_on_holiday_bool_df.columns = ['Total']
tweet_created_on_holiday_bool_df
```

```python
plt.figure(figsize=(8,2))

ax = sns.barplot(x='Total',y=tweet_created_on_holiday_bool_df.index,data=tweet_created_on_holiday_bool_df,orient='h')
ax.set_title('Tweet done on Holidays',fontsize=14)
ax.set_ylabel('')
ax.set_xlabel('Number of Tweets',fontsize=12)
ax.set_yticklabels(['Not Created'])

for p in ax.patches:
    width = p.get_width()
    ax.text(width+1,p.get_y()+p.get_height()/2,'{:1.0f}'.format(width))
sns.despine()
```
